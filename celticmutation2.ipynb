```python
def ngrams_calculation(text, n):
    n_grams_list = []
    for i in range(len(text)):
       n_grams_list.append(tuple(text[i: i + n]))
    return n_grams_list
```

```python
train_word_tag = {}
for s in train_data:
  for w in s:
    for t in s:
      w=w.lower()
      try:
        try:
          train_word_tag[t][w]+=1
        except:
          train_word_tag[t][w]=1
      except:
        train_word_tag[t]={w:1}
```

```python
train_word_tag
```

```python
#Calculating the emission probabilities using train_word_tag
train_emission_prob={}
for k in train_word_tag.keys():
  train_emission_prob[k]={}
  count = sum(train_word_tag[k].values())
  for k2 in train_word_tag[k].keys():
    train_emission_prob[k][k2]=train_word_tag[k][k2]/count
```

```python
bigram_count = {}
unigram_count = {}
def bigram_counts(tags):
  for i_bigram in ngrams_calculation(tags, 2):
    if i_bigram in bigram_count:
      bigram_count[i_bigram] += 1
    else:
      bigram_count[i_bigram] = 1
  return bigram_count
```

```python
bigram_counts(bigram_tag_file)
```

```python
def unigram_counts(tags):
  for tag in tags:
    if tag in unigram_count:
      unigram_count[tag] += 1
    else:
      unigram_count[tag] = 1
  return unigram_count
```

```python
unigram_counts(bigram_tag_file)
```

```python
from collections import defaultdict
unknown_prob=0.0000000000001
transition_probabilities = defaultdict(lambda: unknown_prob)
def transition_probability(tags):
  bigrams = ngrams_calculation(tags, 2)
  for bigram in bigrams:
    transition_probabilities[bigram] = bigram_count[bigram] / unigram_count[bigram[0]]
  return transition_probabilities
```
